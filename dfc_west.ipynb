```python
from dotenv import load_dotenv
import os

load_dotenv()  # loads variables from .env into environment

mapbox_api_key = os.getenv('MAPBOX_API_KEY')
census_api_key = os.getenv('CENSUS_API_KEY')
```

```python
# Load your geocoded Oregon clinics
df = pd.read_csv("filtered_clinics_oregon.csv")

features = []

for idx, row in df.iterrows():
    lat = row["latitude"]
    lon = row["longitude"]

    url = (
        f"https://api.mapbox.com/isochrone/v1/mapbox/driving/"
        f"{lon},{lat}"
        f"?contours_minutes=30&polygons=true"
        f"&access_token={MAPBOX_API_KEY}"
    )

    resp = requests.get(url)
    if resp.status_code == 200:
        gj = resp.json()
        features.extend(gj["features"])
    else:
        print(f"Failed for row {idx}: {resp.status_code}")

# Save all isochrone polygons into one GeoJSON file
collection = geojson.FeatureCollection(features)

with open("oregon_isochrones.geojson", "w") as f:
    geojson.dump(collection, f)
```
